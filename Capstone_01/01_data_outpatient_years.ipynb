```python
## Relevant modules.
import pandas as pd
import os
```

```python
## Defining range of years
years = list(range(2008, 2013))

count_all_observations = []
for year in years:
    file_in = 'original_data/datos_ambulatorios_'+str(year)+'_new.csv'
    data_outpatient_year = pd.read_csv(file_in, header=0)
    
    data_outpatient_year['year'] = [year]*len(data_outpatient_year)
    col_names = ['doc_name','spec_code','spec_es','n_visits','n_surgeries','n_patients','year']
    data_outpatient_year.columns = col_names
    data_outpatient_year['spec_es'] = [i.lower() for i in data_outpatient_year['spec_es']]
    
    doc_code = []
    for i in data_outpatient_year['doc_name']:
        for j in i.split('-'):
            if j.isdigit():
                doc_code.append(int(j))

    data_outpatient_year['doc_code'] = doc_code
    col_order = ['spec_code','spec_es','year','doc_code','doc_name','n_visits','n_surgeries','n_patients']
    data_outpatient_year = data_outpatient_year[col_order]

    doc_code_ck = [int(j) for i in data_outpatient_year['doc_name'] for j in i.split('-') if j.isdigit()]
    if all(data_outpatient_year['doc_code'] == doc_code_ck):
        data_outpatient_year.drop('doc_name', axis=1, inplace=True)
    
    data_outpatient_year.sort_values('spec_code', inplace=True)
    data_outpatient_year.reset_index(drop=True, inplace=True)
    count_all_observations.append(data_outpatient_year['doc_code'].count())

    if os.path.exists('data/data_outpatient_'+str(year)+'.csv'):
        print('data/data_outpatient_'+str(year)+'.csv'+' already exists.')
    elif os.path.exists('data'):
        file_out = 'data/data_outpatient_'+str(year)+'.csv'
        data_outpatient_year.to_csv(path_or_buf=file_out, index=False)
        print('data/'+' already exists.')
        print('data/data_outpatient_'+str(year)+'.csv'+' has been created.')
    else:
        os.mkdir('data')
        file_out = 'data/data_outpatient_'+str(year)+'.csv'
        data_outpatient_year.to_csv(path_or_buf=file_out, index=False)
        print('data/data_outpatient_'+str(year)+'.csv'+' has been created.')
    print(data_outpatient_year.info(),'\n') 

print('The total number of observations is', sum(count_all_observations), end='.')
```
